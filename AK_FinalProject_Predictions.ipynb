```python
data = data.drop(columns=['date'])
data
```

```python
#
# let's convert our dataframe to a numpy array, named A
#
A = data.to_numpy()
A = A.astype('float64')    # many types:  www.tutorialspoint.com/numpy/numpy_data_types.htm
print(A[:,:])

# print(A)  # the five rows above is probably enough...
```

```python
#
# nice to have NUM_ROWS and NUM_COLS around
#
NUM_ROWS, NUM_COLS = A.shape
print(f"\nThe dataset has {NUM_ROWS} rows and {NUM_COLS} cols")
```

```python
#print("+++ Start of data definitions +++\n")

#
# we could do this at the data-frame level, too!
#

X_all = A[:,1:5]   # X (features) ... is all rows, columns  1, 2, 3, 4
y_all = A[:,0]    # y (labels) ... is all rows, columns 0

print("First 5 quarters of the Target: ", y_all[:5])
print("First 5 quarters of the Features: ", X_all[:5])
```

```python
#
# we can scramble the data, to remove (potential) dependence on its ordering:
#
indices = np.random.permutation(len(y_all))  # indices is a permutation-list

# we scramble both X and y, necessarily with the same permutation
X_all = X_all[indices]              # we apply the _same_ permutation to each!
y_all = y_all[indices]              # again...

print(X_all)
```

```python
#
# We next separate into test data and training data ...
#    + We will train on the training data...
#    + We will _not_ look at the testing data to build the model
#
# Then, afterward, we will test on the testing data -- and see how well we do!
#

#
# a common convention:  train on 80%, test on 20%    Let's define the TEST_PERCENT
#

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.2)

print(f"training with {len(y_train)} rows;  testing with {len(y_test)} rows\n" )

#print(f"+++ Testing +++   Held-out data... (testing data: {len(y_test)})\n")
#print(f"y_test: {y_test[0:5,:]}\n")
#print(f"X_test (few rows): {X_test[0:5,:]}")  # 5 rows
#print("\n")
#print(f"+++ Training +++   Data used for modeling... (training data: {len(y_train)})\n")
#print(f"y_train: {y_train[0:5,:]}\n")
#print(f"X_train (few rows): {X_train[0:5,:]}")  # 5 rows
```

```python
# Import necessary modules
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error



# Initialize and train the regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("+++ Testing +++   Held-out data... (testing data: {})\n".format(len(y_test)))
print(f"First 5 actual y_test values: {y_test[0:5]}")
print(f"First 5 predicted y values: {y_pred[0:5]}\n")
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")

# Optional: Print the coefficients and intercept
print(f"Coefficients: {regressor.coef_}")
print(f"Intercept: {regressor.intercept_}")
```

```python


# Store the results in a DataFrame
results = {
    "y_actual": y_test,
    "y_predicted": y_pred
}

# Create a DataFrame
results_df = pd.DataFrame(results)

print(results_df)
print()
print()

# Store the results in a dictionary
results1 = {
    "Metric": ["Mean Squared Error (MSE)", "R² Score", "Mean Absolute Error (MAE)"],
    "Value": [mse, r2, mae]
}

# Create a DataFrame
results1_df = pd.DataFrame(results1)

# Display the results as a table
print(results1_df)


# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(results_df.index, results_df['y_actual'], label='Actual', color='blue', marker='o')
plt.plot(results_df.index, results_df['y_predicted'], label='Predicted', color='red', linestyle='--', marker='x')
plt.fill_between(results_df.index, results_df['y_actual'], results_df['y_predicted'], color='gray', alpha=0.2)

plt.title('Actual vs Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Fetch Microsoft stock data
msft_data = yf.download('MSFT', start='2020-01-01', end='2023-10-31')

# Display the first few rows of the data
print(msft_data.head())
```

```python
# Prepare the data
msft_data['Date'] = msft_data.index
msft_data = msft_data[['Date', 'Close']]

# Create lag features using .loc to avoid SettingWithCopyWarning
for lag in range(1, 11):
    msft_data[f'Lag_{lag}'] = msft_data['Close'].shift(lag)
msft_data = msft_data.dropna()

# Prepare features and target
X = msft_data[[f'Lag_{lag}' for lag in range(1, 11)]]
y = msft_data['Close']

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(X_train.head())
print(y_train.head())
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

# Convert the data into DMatrix format for XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Set parameters for XGBoost
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'eta': 0.01,
    'max_depth': 5,  # Increased max_depth for better learning
    'subsample': 0.8,  # Adding subsample for regularization
    'colsample_bytree': 0.8  # Adding colsample_bytree for regularization
}

# Train the model
num_boost_round = 2000  # Increased number of boosting rounds
early_stopping_rounds = 50  # Set early stopping rounds to avoid overfitting
model = xgb.train(params, dtrain, num_boost_round, evals=[(dtrain, 'train'), (dtest, 'test')], early_stopping_rounds=early_stopping_rounds, verbose_eval=False)

# Make predictions
y_pred = model.predict(dtest)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")

# Optional: Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(msft_data['Date'][-len(y_test):], y_test, label='Actual', color='blue', marker='o')
plt.plot(msft_data['Date'][-len(y_test):], y_pred, label='Predicted', color='red', linestyle='--', marker='x')
plt.fill_between(msft_data['Date'][-len(y_test):], y_test, y_pred, color='gray', alpha=0.2)

plt.title('Actual vs Predicted Microsoft Stock Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```
